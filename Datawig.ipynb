```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import mean_squared_error
from math import sqrt
import random
import numpy as np
random.seed(0)
#Fetching the dataset
import pandas as pd
```

```python
data=pd.read_csv('diabetes.csv')
```

```python
data.head()
```

```python
#Randomly replace 40% of the first column with NaN values
column=data['PlasmaGlucose'];
print(column.size)
```

```python
missing_pct=int(column.size*0.4)
i=[random.choice(range(column.shape[0]))for _ in range(missing_pct)]
column[i]=np.NaN
print(column.shape[0])
```

```python
column.head()
```

```python
import datawig
```

```python
imputer = datawig.SimpleImputer(
    input_columns=['Age'], # column(s) containing information about the column we want to impute
    output_column= 'PlasmaGlucose', # the column we'd like to impute values for
    output_path = 'imputer_model' # stores model data and metrics
    )
```

```python
#Fit an imputer model on the train data
imputer.fit(data)
data = imputer.predict(data)
data['PlasmaGlucose'] = data['PlasmaGlucose_imputed']
del data['PlasmaGlucose_imputed']
```

```python
data
```

```python
# Convert dataset into numpy array
X = (data.iloc[:, :-1]).values
Y = (data.iloc[:,-1]).values
```

```python
#After data preprocessing 
data1=pd.DataFrame(X)
data1
```

```python
#actual values of diabetes
data2=pd.DataFrame(Y)
data2
```

```python
#importing libraries for applying classification algorithm 'Decision Tree'
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=1) # 70% training and 30% test
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#predicted values of Diabetes
y_pred=pd.DataFrame(y_pred)
y_pred
```
